```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for unique buyers and their quantities
buyer_total_quantity = df.groupby('Buyer')['Quantity'].sum().reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=buyer_total_quantity, x='Buyer', y='Quantity')
plt.title('Total Quantity per Buyer')

```

```python
# Find the maximum quantity bought by each buyer
max_quantity_by_buyer = df.groupby('Buyer')['Quantity'].max().reset_index()

# Display the resulting DataFrame
print(max_quantity_by_buyer)
```
